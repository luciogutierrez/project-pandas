```python
# Chosing relevants columns from main data
def select_useful_columns_1(df):
    df_ = df.copy()
    df_ = df_[['Case Number', 'Type', 'Country', 'Activity', 'Sex ', 'Age', 'Injury', 'Fatal (Y/N)', 'Species ']]
    # data.head(5)  
    return df_
```

```python
# Change some titles names
def rename_columns(df):
    df_ = df.copy()
    df_.columns = ['Case Number', 'Type', 'Country','Activity', 'Sex', 'Age', 'Injury', 'Fatal', 'Species']
    # data.head(5)    
    return df_
```

```python
# null's treatment
def data_nulls(df):
    df_ = df.copy()
    df_.dropna(axis=1, how='all', inplace=True)
    df_.dropna(axis=0, how='all', inplace=True)

    for col in df_._get_numeric_data():
        df_[col] = df_[col].fillna(0)

    main_cols = ['Type','Sex', 'Activity', 'Sex', 'Injury','Fatal','Species']
    for col in main_cols:
        df_[col] = df_[col].fillna('Unidentified')

    return df_
```

```python
# numeric columns treatment
def to_numeric(df, col):
    df_ = df.copy()
    df_[col] = pd.to_numeric(df_[col], errors='coerce')
    df_[col] = df_[col].fillna(0)
    return df_
```

```python
# return just numeric values in string
def get_nums(nums):
    num = str(nums)[:2]
    return ''.join(n for n in str(num) if n in digits)
```

```python
# get numeric values from data
def get_numeric(df, col):
    df_ = df.copy()
    df_[col] = df_[col].apply(get_nums)
    return df_
```

```python
# fill NaN with an specific value
def fill_na(df, col, val):
    df_ = df.copy()
    df_[col] = df_[col].fillna(val)

    if df_[col].str.strip().empty:
        df_[col] = '-'

    return df_
```

```python
# Change the type of a column
def cast_colum(df, col, type_):
    df_ = df.copy()
    df_[col] = df_[col].astype(type_)

    return df_
```

```python
# Extract date from a object
def extract_date(date_obj):
    str_date = str(date_obj)
    año = str_date[:4]
    mes = str_date[5:7]
    dia = '01'
    sec = '00:00:00'
    strDate = año +'-'+ mes +'-'+ dia +' '+ sec
    try:
        date_time = datetime.strptime(strDate, '%Y-%m-%d %H:%M:%S')    
    except:
        strDate = '1900-01-01 00:00:00'
        date_time = datetime.strptime(strDate, '%Y-%m-%d %H:%M:%S')    

    return date_time
```

```python
# Get date from field with date in
def get_date(df, col):
    df_ = df.copy()
    df_['Date'] = df_[col].apply(extract_date)
    return df_
```

```python
# Select final columns
def select_useful_columns_2(df):
    df_ = df.copy()
    df_= df_[['Type', 'Country', 'Activity', 'Sex', 'Age', 'Injury', 'Fatal', 'Species', 'Date']]
    # data.head(5)  
    return df_
```

```python
# to arrange columns appropriately
def re_order_columns(df):
    df_ = df.copy()
    col_order = ['Date', 'Type', 'Country', 'Activity', 'Sex', 'Age', 'Injury', 'Fatal', 'Species']
    df_ = df_[col_order]
    # data.head(5)    
    return df_
```

```python
# Main Pipeline 
df1 = select_useful_columns_1(data)
df2 = rename_columns(df1)
del df1
df3 = data_nulls(df2)
del df2
df4 = to_numeric(df3, 'Age')
del df3
df5 = get_numeric(df4, 'Age')
del df4
df6 = get_date(df5,'Case Number')
del df5
df7 = fill_na(df6, 'Age', 0)
del df6
df8 = select_useful_columns_2(df7)
del df7
df9 = re_order_columns(df8)
del df8
df9.head()
```

```python
# Save dataset into a csv file.
df9.to_csv('shark_attack_clean_dataset.csv')
```
